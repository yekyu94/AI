## 2. 추출

```python
c = df['동반자']
```

```python
matched = re.match(r'(\d+)명', c[0])
```

```python
matched.group(0)
```

```python
matched.group(1)
```

```python
int(matched.group(1))
```

```python
total = 0
for i in c:
    matched = re.match(r'(\d+)명', i)
    num = int(matched.group(1))
    total += num
total
```

## 3. 교체

```python
comp = re.compile(r'\((?:(?P<hour>\d+)h ?)?(?:(?P<minute>\d+)m)?\)')
comp
```

```python
search = comp.search(df['담당'][2])
search
```

```python
search.groupdict()
```

```python
search.group('hour')
```

```python
search.group('minute')
```

```python
re.sub('(\d+)h (\d+)m', '\g<1>시간 \g<2>분', '1h 20m')
```

```python
re.sub('(?P<hour>\d+)h (?P<minute>\d+)m', '\g<hour>시간 \g<minute>분', '1h 20m')
```
